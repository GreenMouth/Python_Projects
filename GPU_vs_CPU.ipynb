```python
#Import required lib
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
#Define number of arrays; Increase array shape linearly by 1
value_range = list(range(1,1001,1))
print('Number of Matrix multiplication done will be', len(value_range))
print('Note: In each consecutive Matrix calculation, size of the Matrix will be linearly increased by 1')
```

```python
start = time.time()

run_time_ms = []
metric_size = []

for value in value_range:
    a = np.full((value,value),5)  
    metric_size.append(value)
    start_time = time.time()
    c = a*a
    end_time = time.time()
    total_time = (end_time - start_time)*1000
    run_time_ms.append(total_time)

end = time.time()
total_time_all_cal =  (end - start)
#print('CPU: Total Time taken for', len(value_range), 'calculations:', total_time_all_cal*1000, 'Millisecond')
print('CPU: Total Time taken for', len(value_range), 'calculations:', total_time_all_cal, 'Second')
```

```python
#Plot 'Multiplication Run time' vs 'Size of Matrix'
plt.plot(value_range, run_time_ms)
plt.grid(axis='y', linestyle='-')
plt.title('Multiplication_Run_Time vs Size_of_Matrix')
plt.xlabel('Size_of_Matrix')
plt.ylabel('Run_Time_ms')
plt.show()
#Note: As size of Matrix is increasing means #calculations are increasing
```
